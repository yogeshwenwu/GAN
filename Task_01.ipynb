#Task_01: Predict Restaurant Ratings

###Objective:
> *Build a machine learning model to predict the aggregate rating of a restaurant based on other features.*

````python
import pandas as pd
````

````python
df = pd.read_csv("Dataset.csv")
df.head()
````

###Step_01 : Preprocessing the Dataset

####==>  *Handling missing values*

````python
df.info()
````

``` 9  | Cuisines | 9542 | non-null |   object```

Column Cuisines has **9** missing values  

````python
df['Cuisines'].fillna(value='unknown',inplace=True)
df['Cuisines'].info()
````

>
####==> Dimensionality Reduction



````python
x = df.drop(['Restaurant ID','Restaurant Name','Country Code','City', 'Address', 'Locality', 'Locality Verbose', 'Rating color', 'Aggregate rating'], axis=1)
y = df['Aggregate rating']
x
````

>
####==>  Encoding Categorical Variables

````python
from sklearn.preprocessing import LabelEncoder

encoder = LabelEncoder()

for column in x:
    x[column] = encoder.fit_transform(x[column])
````

````python
x.head()
````

>
####==>  Splitting the data into training and testing sets

````python
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(x,y,test_size=0.3,random_state=42)
````

````python
print("x_train :\n",x_train)
print()
print("y_train :\n",y_train)
````

###Step_02 : Select a Regression Algorithm

*(e.g., linear regression,decision tree regression) and train it on the training data.*

````python
from sklearn.ensemble import RandomForestRegressor

reg = RandomForestRegressor()
````

````python
reg.fit(x_train, y_train)
````

###Step_03 : Evaluate the model's performance using appropriateregression metrics

````python
from sklearn.metrics import mean_squared_error, r2_score
````

````python
y_hat = reg.predict(x_test)
````

####==> *Mean Squared Error on the testing data.*

````python
mse = mean_squared_error(y_test, y_hat)

print(f'Mean Squared Error :  {mse}')
````

####==> *R-squared on the testing data.*

````python
r2_scr = r2_score(y_test, y_hat)

print(f'R-Squared:  {r2_scr}')
````

###Step_04 : Interpret the model's results and analyze the mostinfluential features affecting restaurant ratings.

````python
ftr_imp = reg.feature_importances_

column = x_train.columns

imp_ftr = pd.DataFrame({'Feature': column, 'Importance':ftr_imp})

imp_ftr_srt = imp_ftr.sort_values(by='Importance', ascending = False)
print(imp_ftr_srt)
````
